#Data preprocessing

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['Heart Disease' ],axis=1)
y=df['Heart Disease']
```

```python
#The dataset is split into training and test set
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

#Building and evaluation of ML models

```python
def report(actual,pred) :
    acc_score=accuracy_score(actual, pred)
    con_mat=confusion_matrix(actual, pred )
    class_rep=classification_report(actual,pred) #To print precision, recall and F1 score
    print("Accuracy of model is", round((acc_score*100),2),'%')
    print( '\n',con_mat)
    print( '\n' ,class_rep)
```

#Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
log.fit(xtrain,ytrain) #Training the model
train_log=log.predict (xtrain) #Predictions on train data
test_log=log. predict (xtest) #Predictions on test data
report(ytrain,train_log)
report(ytest, test_log)
```

```python
#Plotting confusion matrix
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
ConfusionMatrixDisplay.from_estimator(log,xtest,ytest)
```

```python
#Plotting ROC curve
from sklearn.metrics import RocCurveDisplay
svc_disp = RocCurveDisplay.from_estimator(log, xtest, ytest)
rfc_disp = RocCurveDisplay.from_estimator(log, xtest, ytest, ax=svc_disp.ax_)
```

#Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier( n_estimators=15, max_depth=4, min_samples_split=10,random_state=1000)
#n_estimators - Number of trees in the forest
#max_depth - Maximum depth of each decision tree in the forest
#min_samples_split - This parameter sets the minimum number of samples required to split an internal node. If the number of samples at a node is below this threshold, the node will not be split because number of samples is less, and it will become a leaf node
rf.fit(xtrain,ytrain)
train_rf=rf.predict(xtrain)
test_rf=rf.predict (xtest)
report(ytrain,train_rf)
report (ytest, test_rf)
```

```python
#Plotting confusion matrix
ConfusionMatrixDisplay.from_estimator(rf,xtest,ytest)
```

```python
#Plotting ROC curve
from sklearn.metrics import RocCurveDisplay
svc_disp = RocCurveDisplay.from_estimator(rf, xtest, ytest)
rfc_disp = RocCurveDisplay.from_estimator(rf, xtest, ytest, ax=svc_disp.ax_)
```

#Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
#max_depth : Max depth of the tree
#The decision tree is split based on Gini impurity criterion. Ranges from 0 to 1 - 0 implies all elements belong to single class ; 1 implies the elements are equally split between both the classes
#random_state = 0 implies that each time the code is run, the same value is displayed
```

```python
dt.fit(xtrain, ytrain)
```

```python
train_dt=dt.predict (xtrain)
test_dt=dt. predict (xtest)
report(ytrain,train_dt)
report (ytest, test_dt)
```

```python
#Plotting confusion matrix
ConfusionMatrixDisplay.from_estimator(dt,xtest,ytest)
```

```python
#Plotting ROC curve
from sklearn.metrics import RocCurveDisplay
svc_disp = RocCurveDisplay.from_estimator(dt, xtest, ytest)
rfc_disp = RocCurveDisplay.from_estimator(dt, xtest, ytest, ax=svc_disp.ax_)
```
